##### Download data from the given source

```python
# Functin to download file from source
def file_exist(file_name):
    return file_name in os.listdir(os.getcwd())
def download_file(url,file_name):
    if file_exist(file_name):
        return
    response = requests.get(url)
    open(file_name, "wb").write(response.content)
```

```python
# Load the dataset1 from source
url = "https://drive.google.com/file/d/1LNcduA_A0G6UZa8TZj-kjDykGa4mA_tC/view?usp=sharing"
download_file(url,"data1.csv")
```
